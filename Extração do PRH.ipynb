## Script para extração de informação dos arquivos do PRH

```python
import pandas as pd
import datetime
from pathlib import Path
import os
```

```python
momentoInicial = datetime.datetime.now() # Inicia um contador de tempo

# Define os arquivos de entrada e saída
extensao = "*.txt"
pastaentrada =  "PRH"
pastaArquivosCorpus = Path(pastaentrada).glob(extensao)

# Iterando os arquivos da pasta de entrada
nome_arquivo = []
texto_arquivo = []

for path in pastaArquivosCorpus: 
    arquivo = os.path.basename(path) 
    nome_arquivo.append(arquivo)
    print("Lendo o arquivo ", arquivo)
    
    texto = ''
    with open(path, 'r', encoding="UTF-8") as f:
        texto = f.read() #Lendo os arquivos da pasta de entrada
        texto = texto.replace('\n','')
        texto_arquivo.append(texto[:300])
```

```python
lista_PRH = {'nome_arquivo': nome_arquivo, 'texto_arquivo': texto_arquivo}
lista_PRH = pd.DataFrame(data=lista_PRH)
lista_PRH.to_csv("lista_PRH.csv", sep=',')
lista_PRH
```
